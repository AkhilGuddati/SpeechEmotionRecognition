```python
import pandas as pd
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd

audiopath = pd.read_csv('./datapath0.csv')
audiopath = np.array(audiopath)
```

```python
def emote(i):
    print("Emotion:",audiopath[i][1].upper())
```

```python
def plt_waveplot(i):
  audio_wav=audiopath[i][0]
  audio,sr=librosa.load(audio_wav)
  #print("File_Path:",audio_wav)
  #print("Sampling_Rate:",sr)
  plt.figure(figsize=(25,6)) 
  librosa.display.waveshow(audio,alpha=0.5)
  plt.xlabel('Time')
  plt.ylabel('Amplitude')
  plt.show()
```

```python
def frequency_spectrum(i):
  audio_wav=audiopath[i][0]
  audio,sr=librosa.load(audio_wav)
  ft_audio = np.fft.fft(audio)
  y_audio = np.abs(ft_audio)
  frequency = np.linspace(0, sr, len(y_audio))
  num_frequency_bins = int(len(frequency)*0.5)
  #print("File_Path:",audio_wav)
  #print("Sampling_Rate:",sr)
  plt.figure(figsize=(25,6))
  plt.plot(frequency[:num_frequency_bins] , y_audio[:num_frequency_bins])
  plt.xlabel("Frequency (Hz)")
  plt.xlabel("Magnitude")
  plt.show()
```

```python
def mel_spectrpgram(i):
  audio_wav=audiopath[i][0]
  audio,sr=librosa.load(audio_wav)
  mel_audio = librosa.feature.melspectrogram(audio, sr=sr, n_fft=2048, hop_length=512, n_mels=20)
  log_mel_audio = librosa.power_to_db(mel_audio)
  plt.figure(figsize=(25,6))
  librosa.display.specshow(log_mel_audio,x_axis="time",y_axis="mel",sr=sr)
  plt.colorbar(format="%+2.f")
  plt.show()
```

```python
emote(11792)
plt_waveplot(11792)
frequency_spectrum(11792)
ipd.Audio(audiopath[11792][0])
mel_spectrpgram(11792)
```
