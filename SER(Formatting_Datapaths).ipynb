```python
from os import listdir
from os.path import isfile, join
import pandas

path = "./audio/TESS Toronto emotional speech set data/"     # Path to folder with TESS audio files

l=listdir(path)
paths=[]
emotions=[]

for k in l:
  onlyfiles = [f for f in listdir(path+k+"/") if 
               isfile(join(path+k+"/", f))]     # Finding the path to audio files
  
  for i in onlyfiles:
    if "pleasant_surprised" in k[4:].lower():
        emotions.append("surprised")    # Extracting labels
    else:
        emotions.append(k[4:].lower())    # Extracting labels
    paths.append(path+k+"/"+i) 

dictionary={"Path":paths,"Label":emotions}
dataset1=pandas.DataFrame(dictionary)
dataset1.to_csv("./datapath1.csv", index=False)
dataset1
```

```python
from os import listdir
from os.path import isfile, join
import pandas

path = "./Audio/ALL"   # Path to folder with SAVEE audio

paths=[]
emotions=[]

# Finding the path to audio files
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]

# Extracting labels
for i in onlyfiles:
    if "_h" in i:
      emotions.append("happy")
    elif "_a" in i:
      emotions.append("angry")
    elif "_d" in i:
      emotions.append("disgust")
    elif "_f" in i:
      emotions.append("fear") 
    elif "_su" in i:
      emotions.append("surprised")
    elif "_sa" in i:
      emotions.append("sad")      
    elif "_n" in i:
      emotions.append("neutral")  
    paths.append(path+i)

dictionary={"Path":paths,"Label":emotions}
dataset2=pandas.DataFrame(dictionary)
dataset2.to_csv("./datapath2.csv", index=False)
dataset2
```

```python
from os import listdir
from os.path import isfile, join
import pandas

path = "./Audio/archive/"   # Path to folder with RAVDESS audio

l=listdir(path)
paths=[]
emotions=[]

for k in l:
  # Finding the path to audio files  
  onlyfiles = [f for f in listdir(path+k+"/") if isfile(join(path+k+"/", f))]
  
  # Extracting labels
  for i in onlyfiles:
    if i[6:8]=="01":
        emotions.append("neutral")
    elif i[6:8]=="02":
        emotions.append("calm")
    elif i[6:8]=="03":
        emotions.append("happy")
    elif i[6:8]=="04":
        emotions.append("sad")
    elif i[6:8]=="05":
        emotions.append("angry")
    elif i[6:8]=="06":
        emotions.append("fear")
    elif i[6:8]=="07":
        emotions.append("disgust")   
    elif i[6:8]=="08":
        emotions.append("surprised")         
    paths.append(path+k+"/"+i)

dictionary={"Path":paths,"Label":emotions}
dataset4=pandas.DataFrame(dictionary)
dataset4.to_csv("./datapath4.csv", index=False)
dataset4
```
